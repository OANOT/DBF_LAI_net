```python
import laspy
import numpy as np

# 读取你的 LAS 文件
file_path = "你的文件路径.las"
las = laspy.read('../las/20230328/20230328_las4_FAPAR.las')

# 1. 查看所有的特征名称（维度）
print("文件包含的特征维度有：")
print(list(las.point_format.dimension_names))



# 3. 检查是否真的有颜色信息（有时候维度里有 red，但数值全是 0）
if hasattr(las, 'intensity'):
    print(f"检测到 RGB 信息。红色通道最大值: {np.max(las.intensity)}")
else:
    print("未检测到 RGB 信息。")

# 4. 查看点数
print(f"总点数: {len(las.points)}")
```
